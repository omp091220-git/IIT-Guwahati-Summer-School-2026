**Problem-4**: Count the number of vowels and consonants in a string.

```python
str1 = input("Enter a sentence: ")
vowels = 0
consonants = 0
for i in str1:
  if i in ['a', 'e', 'i', 'o', 'u'] or i in ['A', 'E', 'I', 'O', 'U']:
    vowels+=1
  else:
    consonants+=1
print("Number of vowels: ", vowels,"\n","Number of consonants: ", consonants)
```

**Problem-13**: Check whether two strings are equal (without using ==).

```python
str1 = input("Enter a sentence: ")
str2 = input("Enter another sentence: ")
if len(str1) != len(str2):
    print("Not Equal")
else:
    equal = True
    for i in range(len(str1)):
        if str1[i] != str2[i]:
            equal = False
            break
    if equal:
        print("Equal")
    else:
        print("Not Equal")
```

**Problem-14**: Count the number of digits, alphabets, and special characters in a string.

```python
str1 = input("Enter a sentence: ")
digits = 0
alphabets = 0
special = 0
for i in str1:
    if i.isdigit():
        digits += 1
    elif i.isalpha():
        alphabets += 1
    else:
        special += 1

print("Digits:", digits)
print("Alphabets:", alphabets)
print("Special Characters:", special)
```

**Problem-15**: Find the longest word in a string.

```python
str1 = input("Enter a sentence: ")
words = str1.split()
longest_word = ""
for word in words:
    if len(word) > len(longest_word):
        longest_word = word
print("Longest word:", longest_word)
```
