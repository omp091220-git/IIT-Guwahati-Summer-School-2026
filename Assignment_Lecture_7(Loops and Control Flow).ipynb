**Problem-10**: Reverse a number

```python
n = int(input("Enter number: "))
rev = 0
while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n //= 10
print(rev)
```

**Problem-11**: Check palindrome number

```python
n = int(input("Enter number: "))
temp = n
rev = 0
while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n //= 10
if temp == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

**Problem-12**: Check Armstrong number

```python
n = int(input("Enter number: "))
temp = n
digits = len(str(n))
sum = 0
while temp > 0:
    digit = temp % 10
    sum += digit ** digits
    temp //= 10
if sum == n:
    print("Armstrong")
else:
    print("Not Armstrong")
```

**Problem-13**: Armstrong numbers in an interval

```python
start = int(input("Start: "))
end = int(input("End: "))
for num in range(start, end + 1):
    digits = len(str(num))
    temp = num
    sum = 0
    while temp > 0:
        digit = temp % 10
        sum += digit ** digits
        temp //= 10
    if sum == num:
        print(num)
```

**Problem-14**: GCD and LCM of two numbers

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
x = a
y = b
while y != 0:
    x, y = y, x % y
gcd = x
lcm = (a * b) // gcd
print("GCD =", gcd)
print("LCM =", lcm)
```

**Problem-15**: Check prime number

```python
n = int(input("Enter number: "))
if n < 2:
    print("Not Prime")
else:
    prime = True
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            prime = False
            break
    if prime:
        print("Prime")
    else:
        print("Not Prime")
```

**Problem-16**: Print prime numbers up to n

```python
n = int(input("Enter n: "))
for num in range(2, n + 1):
    prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            prime = False
            break
    if prime:
        print(num)
```

**Problem-17**: Print prime numbers from m to n

```python
m = int(input("Enter start: "))
n = int(input("Enter end: "))
for num in range(m, n + 1):
    if num > 1:
        prime = True
        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                prime = False
                break
        if prime:
            print(num)
```

**Problem-19**: Replace all 0's with 1's in a number

```python
n = int(input("Enter number: "))
if n == 0:
    print(1)
else:
    result = 0
    place = 1
    while n > 0:
        digit = n % 10
        if digit == 0:
            digit = 1
        result += digit * place
        place *= 10
        n //= 10
    print(result)
```

**Problem-20**: Convert binary to decimal using while

```python
binary = int(input("Enter binary number: "))
decimal = 0
power = 0
while binary > 0:
    digit = binary % 10
    decimal += digit * (2 ** power)
    power += 1
    binary //= 10
print(decimal)
```
