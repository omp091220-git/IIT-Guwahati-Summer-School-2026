# **Demonstration Video Problems**

**Problem-1**:

```python
with open("hello.txt", "w") as f:
  f.write("Hello, this is my first file!\n")
  f.write("File handling in Python is easy.")
```

**Problem-2**:

```python
with open("hello.txt", "a") as f:
  f.write("\nThis line is appended.")
```

**Problem-3**:

```python
with open("hello.txt", "r") as f:
  data = f.read()
  print(data)
```

**Problem-4**:

```python
with open("hello.txt", "r") as f:
  for line in f:
    print(line.strip())
```

**Problem-5**:

```python
with open("hello.txt", "r") as f:
  data = f.read()
words = data.split()
print("Total words: ", len(words))
```

**Problem-6**:

```python
with open("hello.txt", "r") as fsrc:
  with open("copy.txt", "w") as fdst:
    for line in fsrc:
      fdst.write(line)
```

**Problem-7**:

```python
import csv
data = [["Name", "Age"], ["Alice", 21], ["Bob", 22]]
with open("students.csv", "w", newline="") as f:
  writer = csv.writer(f)
  writer.writerows(data)
with open("students.csv", "r") as f:
  reader = csv.reader(f)
  for row in reader:
    print(row)
```

**Problem-8**:

```python
import pickle
data = {"name": "Alice", "age": 25}
with open("data.pkl", "wb") as f:
  pickle.dump(data, f)
with open("data.pkl", "rb") as f:
  loaded_data = pickle.load(f)
print(loaded_data)
```

**Problem-9**:

```python
with open("hello.txt", "r") as f:
  print(f.tell())
  print(f.read(5))
  print(f.tell())
  f.seek(0)
  print(f.tell())
```
