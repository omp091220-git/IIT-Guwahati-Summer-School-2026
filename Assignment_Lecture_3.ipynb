**Problem-1**: Write a Python program to check whether the given number is positive or negative using simple if statement.

```python
num1 = int(input("Enter a number: "))
if num1>=0:
  print(num1, "is positive.")
else:
  print(num1, "is negative.")
```

**Problem-2**: Write a Python program to check whether the given number is even or odd using modulus operator.

```python
num1 = int(input("Enter a number: "))
if num1%2==0:
  print(num1, "is even.")
else:
  print(num1, "is odd.")
```

**Problem-3**: Write a Python program to check whether the given number is even or odd using bitwise operator.

```python
num1 = int(input("Enter first number: "))
if num1&1==0:
  print(num1, "is even.")
else:
  print(num1, "is odd.")
```

**Problem-4**: Write a Python program to check if two numbers are equal or not.

```python
num1 = int(input("Enter a number: "))
num2 = int(input("Enter another number: "))
if num1==num2:
  print(num1, num2, "are equal.")
else:
  print(num1, num2, "are not equal.")
```

**Problem-5**: Write a Python program to check if two numbers are equal or not using bitwise operator.

```python
num1 = int(input("Enter a number: "))
num2 = int(input("Enter another number: "))
if num1|num2==num1:
  print(num1, num2, "are equal.")
else:
  print(num1, num2, "are not equal.")
```

**Problem-6**: Write a Python program to find the largest number among two numbers.

```python
num1 = int(input("Enter a number: "))
num2 = int(input("Enter another number: "))
if num1>=num2:
  print(num1,"is larger than ",num2)
else:
  print(num2, "is larger than ", num1)
```

**Problem-7**: Write a Python program to find the smallest number among two numbers.

```python
num1 = int(input("Enter a number: "))
num2 = int(input("Enter another number: "))
if num1<=num2:
  print(num1,"is smaller than ",num2)
else:
  print(num2, "is smaller than ", num1)
```

**Problem-8**: Write a Python program to find the maximum and minimum of two numbers without using any loop or condition.

```python
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
maxi = (num1 + num2 + abs(num1 - num2))//2
mini = (num1 + num2 - abs(num1 - num2))//2
print("The maximum number is: ", maxi)
print("The minimum number is: ", mini)
```
