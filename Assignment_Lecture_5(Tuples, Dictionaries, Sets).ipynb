## **Tuples**

**Problem-1**: Print all elements of a tuple using a for loop.

```python
tuple1 = (1, 2, 3, 4, 5, 6)
for item in tuple1:
  print(item)
```

**Problem-2**: Print all elements of a tuple using a while loop.

```python
tuple1 = (1, 2, 3, 4, 5, 6)
i=0
while i<len(tuple1):
  print(tuple1[i])
  i+=1
```

**Problem-3**: Find the length of a given tuple.

```python
tuple1 = (1, 2, 3, 4, 5, 6)
print(len(tuple1))
```

**Problem-4**: Find the maximum and minimum elements in a tuple.

```python
tuple1 = (1, 2, 3, 4, 5, 6)
print("Maximum: ", max(tuple1))
print("Minimum: ", min(tuple1))
```

**Problem-5**: Count the number of even and odd elements in a tuple.

```python
tuple1 = (1, 2, 3, 4, 5, 6)
even_count = 0
odd_count = 0
for number in tuple1:
  if number % 2 == 0:
    even_count += 1
  else:
    odd_count += 1
print("Even Count: ", even_count)
print("Odd Count: ", odd_count)
```

**Problem-6**: Convert a tuple into a list.

**Problem-7**: Convert a list into a tuple.

**Problem-8**: Check whether an element exists in a tuple.

**Problem-9**: Find the index of a given element in a tuple.

**Problem-10**: Count the frequency of an element in a tuple.

**Problem-11**: Reverse a tuple.

**Problem-12**: Concatenate two tuples.

**Problem-13**: Check whether a tuple is a palindrome or not.

**Problem-14**: Find the sum of all elements in a numeric tuples.

```python
tuple1 = (1, 2, 3, 4, 5, 6)
sum = 0
for i in tuple1:
  sum += i
print("Sum: ", sum)
```

## **Dictionary**

```python
dict1 = {'fruits': 1, 'vegetables': 2, 'dry fruits': 3}
print("Keys: ", dict1.keys())
print("Values: ", dict1.values())
```

**Problem-2**: Add a new key-value pair to a dictionary.

```python
dict1 = {'fruits': 1, 'vegetables': 2, 'dry fruits': 3}
dict1['spices'] = 4
print(dict1)
```

**Problem-3**: Delete a key from a dictionary.

```python
dict1 = {'fruits': 1, 'vegetables': 2, 'dry fruits': 3}
del dict1['dry fruits']
print(dict1)
```

**Problem-4**: Check whether a key exists in a dictionary.

```python
dict1 = {'fruits': 1, 'vegetables': 2, 'dry fruits': 3}
key = 'dry fruits'
if key in dict1:
  print(key, "exists in the dictionary")
```

**Problem-5**: Find the length of a dictionary.

```python
dict1 = {'fruits': 1, 'vegetables': 2, 'dry fruits': 3}
print(len(dict1))
```

**Problem-6**: Count the frequency of elements in a list using a dictionary.

```python
list1 = ['fruits', 'fruits', 'vegetables', 'dry fruits', 'spices']
frequency = {}
for item in list1:
  if item in frequency:
    frequency[item] += 1
  else:
    frequency[item] = 1
print(frequency)
```

**Problem-7**: Merge two dictionaries into one.

```python
dict1 = {'fruits': 1, 'vegetables': 2}
dict2 = {'dry fruits': 3, 'spices': 4}
dict1.update(dict2)
print(dict1)
```

**Problem-8**: Find the maximum and minimum values in a dictionary.

```python
dict1 = {'fruits': 1, 'vegetables': 2, 'dry fruits': 3}
print("Maximum: ", max(dict1.values()))
print("Minimum: ", min(dict1.values()))
```

**Problem-9**: Print only the keys of a dictionary.

**Problem-10**: Print only the values of a dictionary.

```python
dict1 = {'name': 'Om', 'age': 19, 'city': 'Surat'}
for value in dict1.values():
  print(value)
```

**Problem-11**: Update the value of an existing key.

```python
dict1 = {'fruits': 1, 'vegetables': 2, 'dry fruits': 3}
dict1['vegetables'] = 3
print(dict1)
```

**Problem-12**: Sort a dictionary by keys.

```python
dict1 = {'fruits': 1, 'vegetables': 2, 'dry fruits': 3}
sorted_dict = dict(sorted(dict1.items()))
print(sorted_dict)
```

**Problem-13**: Sort a dictionary by values.

```python
dict1 = {'fruits': 1, 'vegetables': 2, 'dry fruits': 3}
sorted_dict = dict(sorted(dict1.items(), key=lambda x: x[1]))
print(sorted_dict)
```

**Problem-14**: Find the sum of all values in a dictionary.

```python
dict1 = {'fruits': 1, 'vegetables': 2, 'dry fruits': 3}
sum = 0
for i in dict1.values():
  sum += i
print(sum)
```

**Problem-15**: Create a dictionary from two lists (keys and values).

```python
list1 = ['fruits', 'vegetables', 'dry fruits']
list2 = [1, 2, 3]
dict1 = dict(zip(list1, list2))
print(dict1)
```

```python
set1 = {1, 2, 3, 4, 5, 6}
set1.remove(5)
print(set1)
```

**Problem-4**: Check whether an element exists in a set.

```python
set1 = {1, 2, 3, 4, 5, 6}
target = 5
if target in set1:
  print(target, "exists in the set")
```

**Problem-5**: Find the length of a set.

```python
set1 = {1, 2, 3, 4, 5, 6}
print(len(set1))
```

**Problem-6**: Find the union of two sets.

```python
set1 = {1, 2, 3}
set2 = {4, 5, 6}
print(set1.union(set2))
```

**Problem-7**: Find the intersection of two sets.

```python
set1 = {1, 2, 3, 4}
set2 = {4, 5, 6}
intersection_set = set1 & set2
print(intersection_set)
```

**Problem-8**: Find the difference between two sets.

```python
set1 = {1, 2, 3}
set2 = {4, 5, 6}
difference_set = set1 - set2
print(difference_set)
```

```python
set1 = {1, 2, 3, 4}
set2 = {4, 5, 6}
sym_diff = set1 ^ set2
print(sym_diff)
```

**Problem-10**: Check whether one set is a subset of another.

```python
set1 = {1, 2, 3}
set2 = {1, 2, 3, 4, 5, 6}
if set1.issubset(set2):
  print("set1 is a subset of set2")
```

**Problem-11**: Check whether one set is a superset of another.

```python
set1 = {1, 2, 3, 4}
set2 = {1, 2, 3}
if set1.issuperset(set2):
  print("set1 is a superset of set2")
```

**Problem-12**: Convert a list into a set to remove duplicates.

```python
list1 = [1, 2, 3, 4, 5, 5, 6]
set1 = set(list1)
print(set1)
```

**Problem-13**: Find common elements between two sets.

```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}
common = set1.intersection(set2)
print("Common: ", common)
```

**Problem-14**: Check whether two sets are equal.

```python
set1 = {1, 2, 3}
set2 = {3, 2, 1}

if set1 == set2:
  print("Yes")
else:
  print("No")
```

**Problem-15**: Perform set operations on multiple sets.

```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}
set3 = {5, 6, 7}
uni = set1 | set2 | set3
print("UNION: ", uni)

inter = set1 & set2 & set3
print("INTERSECTION: ", inter)

diff = set1 - set2 - set3
print("DIFFERENCE: ", diff)

sym_diff = set1 ^ set2 ^ set3
print("SYMMETRIC DIFFERENCE: ", sym_diff)
```
