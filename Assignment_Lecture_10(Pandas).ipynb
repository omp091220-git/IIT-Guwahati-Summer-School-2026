**Problem-9**:

```python
import pandas as pd
import numpy as np
data = {
    "Name": ["Asha", "Ravi", np.nan, "Noor"],
    "Score": [85, np.nan, 77, 98],
    "City": ["Guwahati", "Jorhat", "Pathsala", np.nan]
  }
df_nan = pd.DataFrame(data)
print(df_nan.isna().sum())
```

**Problem-10**:

```python
filled =  df_nan.copy()
print("Original data:\n", filled)
filled["Score"] = filled["Score"].fillna(filled["Score"].mean())
filled = filled.dropna(subset=["City"])
print("After dropping:\n", filled)
print("Mean:\n")
print(filled.groupby("City")["Score"].mean())
```
